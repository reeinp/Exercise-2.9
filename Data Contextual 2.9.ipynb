```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

file_path = r"C:\Users\reinh\Downloads\Data Eksportir Minyak Mentah Indonesia.xlsx"
df = pd.read_excel(file_path)
df
```

```python
# Visualisasi deret waktu
plt.figure(figsize=(12, 6))
plt.plot(df['Tahun'], df['Ekspor Minyak Mentah'], marker='o', color='b')
plt.xlabel('Tahun')
plt.ylabel('Ekspor Minyak Mentah (dalam ribu ton)')
plt.title('Data Ekspor Minyak Mentah Indonesia')
plt.grid(True)
plt.show()
```

```python
# Uji ADF untuk stasioneritas
adf_result = adfuller(df['Ekspor Minyak Mentah'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')
```

```python
# Plot ACF
plt.figure(figsize=(12, 10))
plot_acf(df['Ekspor Minyak Mentah'], lags=20)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF) of Ekspor Minyak Mentah')
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import acf  # Mengimpor fungsi acf

# Membaca data dari file Excel
file_path = r"C:\Users\reinh\Downloads\Data Eksportir Minyak Mentah Indonesia.xlsx"
df = pd.read_excel(file_path)

# Menghitung ACF
acf_result = acf(df['Ekspor Minyak Mentah'], nlags=30)

# Hitung jumlah observasi
N = len(df['Ekspor Minyak Mentah'])

# Hitung standard error
se = 1 / np.sqrt(N)

# Hitung z-statistics
z_statistics = acf_result / se

# Membuat DataFrame
df_acf = pd.DataFrame({'Lags': range(1, len(acf_result)+1),
                       'Sample ACF': acf_result,
                       'Z Statistics': z_statistics})

# Menampilkan DataFrame
print(df_acf)
```
