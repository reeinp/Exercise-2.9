```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

file_path = r"D:\Semester 3\Data Sunspot Number.xlsx"
df = pd.read_excel(file_path)
```

```python
# 1. Plot deret waktu
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Sunspot_Number'], label='Sunspot Number')
plt.xlabel('Year')
plt.ylabel('Sunspot Number')
plt.title('Annual International Sunspot Numbers')
plt.legend()
plt.show()
```

```python
# 2. Hitung ADF
adf_result = adfuller(df['Sunspot_Number'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')
```

```python
# 3. Plot ACF
plt.figure(figsize=(12, 8))
plot_acf(df['Sunspot_Number'], lags=50)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF) of Sunspot Number')
plt.show()
```

```python
# Menghitung sample autocorrelation function
autocorr_values = df['Sunspot_Number'].autocorr()
autocorr_values
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import acf  # Mengimpor fungsi acf

# Membaca data dari file Excel
file_path = "D:\Semester 3\Data Sunspot Number.xlsx"
df = pd.read_excel(file_path)

# Menghitung ACF
acf_result = acf(df['Sunspot_Number'], nlags=20)

# Hitung jumlah observasi
N = len(df['Sunspot_Number'])

# Hitung standard error
se = 1 / np.sqrt(N)

# Hitung z-statistics
z_statistics = acf_result / se

# Membuat DataFrame
df_acf = pd.DataFrame({'Lags': range(1, len(acf_result)+1),
                       'Sample ACF': acf_result,
                       'Z Statistics': z_statistics})

# Menampilkan DataFrame
print(df_acf)
```
